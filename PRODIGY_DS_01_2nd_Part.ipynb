```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the dataset
data = pd.read_csv('/content/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_5871594.csv')
```

```python
data
```

```python
gender_counts = data['Region'].value_counts()
bar_width = 0.9
x=range(len(gender_counts.index))


plt.bar(gender_counts.index,gender_counts.values)
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Distribution of Region')


plt.xticks(x,gender_counts.index,rotation=45)
plt.tight_layout()
plt.show()
```
